```python
import pandas as pd

df=pd.read_csv("D:/Atte1.csv")
import cv2
face_cascade = cv2.CascadeClassifier("C:/Users/91845/Downloads/haarcascade_frontalface_default.xml")
cam = cv2.VideoCapture(0)
Id=input("Enter the data")
Name=input("Enter the name")
df2=pd.DataFrame({"Id":[Id],"Name":[Name]})
df=pd.concat([df,df2]).drop_duplicates().reset_index(drop=True)
df.to_csv("D:/Atte1.csv",index=False)

sampleNum = 0
while(True):
    ret,img=cam.read()
    faces=face_cascade.detectMultiScale(img, 1.3, 5)
    for (x, y, h, w) in faces:
        cv2.rectangle(img,(x, y),(x + w,y + h),(144, 1, 1), 3)
        sampleNum=sampleNum + 1
        cv2.imwrite('C:/Users/91845/Pictures/Saved Pictures/faces/user.'+ Id +'.'+ str(sampleNum) + ".jpg" , img[y:y + h,x:x + w])
        cv2.imshow('frame',img)
    if cv2.waitKey(100) & 0xFF == ord('q'):
        break
    elif sampleNum > 10:
        break
cam.release() 
cv2.destroyAllWindows()

    


```

```python
from PIL import Image
import cv2
import os
import numpy as np
recognizer=cv2.face.LBPHFaceRecognizer_create()
path ="C:/Users/91845/Pictures/Saved Pictures/faces"

def getImageswithID(path):
  
  imagesPaths=[os.path.join(path,f) for f in os.listdir(path)]
  faces=[]
  IDs=[]
  for imagePath in imagesPaths:
    faceImg=Image.open(imagePath).convert('L')
    faceNb=np.array(faceImg,'uint8')
    ID=int(os.path.split(imagePath)[-1].split('.')[1])
    faces.append(faceNb)
    IDs.append(ID)
    cv2.imshow("training",faceNb)
    cv2.waitKey(100) #for 10 millisecond wait......
  return np.array(IDs),faces


Ids,faces=getImageswithID(path) 
recognizer.train(faces,np.array(Ids))
recognizer.save("C:/Users/91845/Desktop/trainningData (1).yml")
cv2.destroyAllWindows()


```

```python
import cv2
import numpy as np
import pandas as pd
date=input("Enter date-dd/mm/yyyy")
faceDetect=cv2.CascadeClassifier("C:/Users/91845/Downloads/haarcascade_frontalface_default.xml")
cam=cv2.VideoCapture(0)
rec = cv2.face.LBPHFaceRecognizer_create()
rec.read("C:/Users/91845/Desktop/trainningData (1).yml")
id=0
font=cv2.FONT_HERSHEY_COMPLEX_SMALL
df=pd.read_csv('D:/Atte1.csv')
while (True):
    ret,img=cam.read()
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = faceDetect.detectMultiScale(gray, 2, 4)
    for(x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x + w,y + h),(0, 0, 255),2)
        id,conf=rec.predict(gray[y:y + h,x:x + w])
        id1 = df["Name"][df.Id==id]
        df[date][df.Id==id] = "P"
        df.to_csv("D:/Atte1.csv",index=False)
        cv2.putText(img,str(id1),(x,y + h), font, 3, 255)
    cv2.imshow("Face",img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cam.release()
cv2.destroyAllWindows()        
```
